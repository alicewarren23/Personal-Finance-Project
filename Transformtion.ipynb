```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

fx_data = pd.read_csv('fx_data.csv')

print("FX data columns:", fx_data.columns)
print("FX data sample:")
print(fx_data.head())

portfolio = pd.read_csv('flat-portfolio_data.csv', index_col=0, parse_dates=True)
print("\nPortfolio data sample:")
print(portfolio.head().iloc[:, :10])
```

```python

fx_data = pd.read_csv('fx_data.csv', skiprows=3)
fx_data.columns = ['Date', 'Close', 'High', 'Low', 'Open', 'Volume']
fx_data['Date'] = pd.to_datetime(fx_data['Date'])
fx_data.set_index('Date', inplace=True)

fx_data['USD_to_GBP'] = 1 / fx_data['Close']  

print("USD to GBP Conversion Rates:")
print(fx_data['USD_to_GBP'].head())


portfolio = pd.read_csv('flat-portfolio_data.csv', index_col=0, parse_dates=True)
all_columns = portfolio.columns
stock_symbols = set([col.split('_')[0] for col in all_columns])
gbp_stocks = [symbol for symbol in stock_symbols if '.L' in symbol]
usd_stocks = [symbol for symbol in stock_symbols if '.L' not in symbol]

print(f"\nGBP Stocks: {gbp_stocks}")
print(f"USD Stocks: {usd_stocks}")


converted_portfolio = portfolio.copy()

print("\nConverting USD stocks to GBP...")

date_count = len(converted_portfolio.index)
for i, date in enumerate(converted_portfolio.index):
    closest_date = fx_data.index[fx_data.index <= date][-1] if any(fx_data.index <= date) else fx_data.index[0]
    conversion_rate = fx_data.loc[closest_date, 'USD_to_GBP']
    
    if i % 50 == 0 or i == date_count - 1:
        print(f"Processing {i+1}/{date_count} dates. Rate on {date.strftime('%Y-%m-%d')}: 1 USD = {conversion_rate:.4f} GBP")
    
    for stock in usd_stocks:
        stock_columns = [col for col in converted_portfolio.columns if col.startswith(f"{stock}_")]
        
        for col in stock_columns:
            if pd.api.types.is_numeric_dtype(converted_portfolio[col]):
                converted_portfolio.loc[date, col] = portfolio.loc[date, col] * conversion_rate

converted_portfolio.to_csv('converted_portfolio_data.csv')

print("\nConversion completed. Data saved to 'converted_portfolio_data.csv'")

print("\nSample of converted data:")
usd_stock_cols = [col for col in converted_portfolio.columns if col.startswith(f"{usd_stocks[0]}_")][:5]
gbp_stock_cols = [col for col in converted_portfolio.columns if col.startswith(f"{gbp_stocks[0]}_")][:5]

print("\nBefore conversion (USD):")
print(portfolio[usd_stock_cols].head())
print("\nAfter conversion (USD to GBP):")
print(converted_portfolio[usd_stock_cols].head())
```

```python
stock_to_viz = usd_stocks[0] 
close_col = f"{stock_to_viz}_Close"

plt.figure(figsize=(12, 6))
plt.plot(portfolio.index, portfolio[close_col], label=f'{stock_to_viz} (USD)', color='blue', alpha=0.7)
plt.plot(converted_portfolio.index, converted_portfolio[close_col], label=f'{stock_to_viz} (GBP)', color='red', alpha=0.7)
plt.title(f'{stock_to_viz} Stock Prices: USD vs. Converted to GBP')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True, alpha=0.3)

from matplotlib.ticker import FuncFormatter

def currency_formatter(x, pos):
    """Format y-axis labels with appropriate currency symbols"""
    return f'${x:.0f} / £{x*0.75:.0f}' 

plt.gca().yaxis.set_major_formatter(FuncFormatter(currency_formatter))

plt.tight_layout()
plt.show()


plt.figure(figsize=(12, 4))
plt.plot(fx_data.index, fx_data['USD_to_GBP'], label='USD to GBP Rate', color='green')
plt.title('USD to GBP Conversion Rate')
plt.xlabel('Date')
plt.ylabel('1 USD = X GBP')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()


orig_mean = portfolio[close_col].mean()
conv_mean = converted_portfolio[close_col].mean()
pct_diff = (conv_mean - orig_mean) / orig_mean * 100

print(f"Original {stock_to_viz} Average Price (USD): ${orig_mean:.2f}")
print(f"Converted {stock_to_viz} Average Price (GBP): £{conv_mean:.2f}")
print(f"Percentage Difference: {pct_diff:.2f}%")
print(f"Average USD to GBP Rate: {fx_data['USD_to_GBP'].mean():.4f} (1 USD = {fx_data['USD_to_GBP'].mean():.4f} GBP)")

fig, ax1 = plt.subplots(figsize=(12, 6))


ax1.set_xlabel('Date')
ax1.set_ylabel('USD Price', color='blue')
ax1.plot(portfolio.index, portfolio[close_col], color='blue', alpha=0.7, label='USD Price')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.grid(True, alpha=0.3)


ax2 = ax1.twinx()
ax2.set_ylabel('GBP Price', color='red')
ax2.plot(converted_portfolio.index, converted_portfolio[close_col], color='red', alpha=0.7, label='GBP Price')
ax2.tick_params(axis='y', labelcolor='red')


plt.title(f'{stock_to_viz} Stock Prices: USD (left axis) vs. GBP (right axis)')
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='best')

plt.tight_layout()
plt.show()
```
